```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import *
from sklearn.svm import *
from sklearn.metrics import *
```

```python
df_train = pd.read_csv('input/train.csv')
m = df_train.shape[0]
display('Number of samples {m}'.format(m=m))

df_test = pd.read_csv('input/test.csv')
m_test = df_test.shape[0]
display('Number of test {m_test}'.format(m_test=m_test))

df_test.head()
```

```python
categories = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

train_comments = df_train['comment_text']
test_comments = df_test['comment_text']

train_word_vec = TfidfVectorizer(analyzer='word', stop_words='english', lowercase=True, sublinear_tf=True)
x_train = train_word_vec.fit_transform(train_comments)

test_word_vec = TfidfVectorizer(analyzer='word', stop_words='english', lowercase=True, sublinear_tf=True, vocabulary=train_word_vec.vocabulary_)
x_test = test_word_vec.fit_transform(test_comments)
```

```python
display('x_train {shape}'.format(shape=x_train.shape))
display('x_test {shape}'.format(shape=x_test.shape))
```

```python
agg_y_pred = pd.DataFrame()
agg_y_pred['id'] = df_test['id']

for category in categories:
    y_train = df_train[category].as_matrix()
    classifier = LogisticRegression(solver='saga', multi_class='multinomial')
    #fit and predict against the training set
    classifier.fit(X=x_train, y=df_train[category].as_matrix())
    y_pred_train = classifier.predict(x_train)
    acc_train = accuracy_score(y_train, y_pred_train)
    display("accuracy score for training set {category} = {accuracy}".format(category=category, accuracy=acc_train))
    
    #predict agains the test set
    y_pred = classifier.predict(x_test)
    agg_y_pred[category] = y_pred
    
```

```python
display(agg_y_pred.loc[agg_y_pred.toxic==1])

agg_y_pred.to_csv('output/lr_submission.csv', index=False)
```
